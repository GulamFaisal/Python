```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('b20.csv')
```

```python
data.head(10)
```

```python
data.drop(['date'],1,inplace=True)
```

```python
data.head(8)
```

```python
prd_days=30
data['Prediction']=data[['value']].shift(-prd_days)
```

```python
data.head(34)
```

```python
data.tail(8)
```

```python
x=np.array(data.drop(['Prediction'],1))
x=x[:len(data)-prd_days]
print(x)
actual_values=x.T
print(actual_values)
```

```python
y=np.array(data['Prediction'])
y=y[:-prd_days]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
prd_days_array=np.array(data.drop(['Prediction'],1))[-prd_days:]
print(prd_days_array)
```

```python
from sklearn.svm import SVR
svr_rbf=SVR(kernel='rbf',C=1e3,gamma=0.00001)
svr_rbf.fit(x_train,y_train)
```

```python
svr_rbf_confidence=svr_rbf.score(x_test,y_test)
print("svr_rbf accuracy: ",svr_rbf_confidence)
```

```python
svm_prediction=svr_rbf.predict(x_test)
print(svm_prediction)
print()
print(y_test)
plt.scatter(y_test,svm_prediction)
plt.xlabel('Testing Module')
plt.ylabel('Predictions')
#Almost Linear
```

```python
#Almost Linear 
```

```python
svm_prediction=svr_rbf.predict(prd_days_array)
print(svm_prediction)
```

```python
plt.plot(svm_prediction,marker='o',color='green')
plt.xlabel('Days')
plt.ylabel('Price Of BitCoin the Corresponding Day')
```
